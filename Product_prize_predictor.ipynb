```python
import pandas as pd

data = {
    "brand": [
        "Nike","Nike","Nike","Adidas","Adidas","Adidas","Puma","Puma","Puma",
        "Reebok","Reebok","Bata","Bata","Skechers","Skechers",
        "Apple","Apple","Apple","Samsung","Samsung","Samsung",
        "HP","HP","Dell","Dell","Lenovo","Lenovo","Asus","Asus",
        "Sony","Sony","LG","LG","OnePlus","OnePlus",
        "Xiaomi","Xiaomi","Realme","Realme"
    ],
    "category": [
        "Shoes","Shoes","Shoes","Shoes","Shoes","Shoes","Shoes","Shoes","Shoes",
        "Shoes","Shoes","Shoes","Shoes","Shoes","Shoes",
        "Electronics","Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics","Electronics",
        "Electronics","Electronics","Electronics","Electronics","Electronics","Electronics"
    ],
    "rating": [
        4.3,4.5,4.6,4.1,4.2,4.3,3.9,4.0,4.1,
        3.8,4.0,3.7,3.9,4.0,4.2,
        4.7,4.8,4.9,4.4,4.5,4.6,
        4.2,4.3,4.3,4.4,4.1,4.2,
        4.0,4.1,4.4,4.5,4.2,4.3,
        4.5,4.6,4.2,4.3,4.0,4.1
    ],
    "price": [
        4200,4500,4800,4000,4200,4400,3600,3800,4000,
        3500,3700,2400,2600,4600,4900,
        78000,82000,86000,65000,69000,72000,
        60000,63000,67000,69000,61000,64000,
        58000,60000,70000,73000,55000,58000,
        52000,54000,18000,19000,20000,21000
    ]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

brand_encoder = LabelEncoder()
category_encoder = LabelEncoder()

df["brand"] = brand_encoder.fit_transform(df["brand"])
df["category"] = category_encoder.fit_transform(df["category"])

df
```

```python
X = df[["brand", "category", "rating"]]   # Input features
y = df["price"]                            # Output
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predicted_prices = model.predict(X_test)
predicted_prices
```

```python
comparison = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": predicted_prices
})

comparison
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(y_test, predicted_prices)
print("Average Error:", error)
```

```python
new_product = pd.DataFrame({
    "brand": [brand_encoder.transform(["Nike"])[0]],
    "category": [category_encoder.transform(["Shoes"])[0]],
    "rating": [4.6]
})

predicted_price = model.predict(new_product)
print("Predicted Price:", predicted_price[0])
```
